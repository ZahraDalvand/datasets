```python
import cv2
from matplotlib import pyplot as plt
```

```python
face_model = cv2.CascadeClassifier("aiolearn.xml")
```

```python
image = cv2.imread("ana.jpg")
```

```python
gray = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
plt.imshow(image)
image2 = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.imshow(image2)
```

```python
faces = face_model.detectMultiScale(
    gray,
    scaleFactor = 1.1 ,
    minNeighbors = 5 ,
    minSize = (30 , 30)
     
)
```

```python
print("Tedad face in image :" , len(faces))
```

```python
for (x ,y ,w ,h) in faces :
    cv2.rectangle(image2 , (x,y), (x+w , y+h), (255 , 255 , 255) , 3)
```
